```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.preprocessing import OneHotEncoder

# Generate dataset
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2,
                           n_informative=10, n_redundant=0, random_state=42)
y = y.reshape(-1, 1)

# One-hot encode labels for softmax
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

# Normalize features
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
X_train = (X - mean) / std
y_train = y

# Activation functions
def ReLU(x):
    return np.maximum(0, x)

def ReLU_derivative(x):
    return np.where(x > 0, 1, 0)

def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

# Initialize weights
def init_weights(input_size, hidden1_size, hidden2_size, output_size):
    np.random.seed(42)
    W1 = np.random.randn(input_size, hidden1_size) * 0.01
    b1 = np.zeros((1, hidden1_size))
    W2 = np.random.randn(hidden1_size, hidden2_size) * 0.01
    b2 = np.zeros((1, hidden2_size))
    W3 = np.random.randn(hidden2_size, output_size) * 0.01
    b3 = np.zeros((1, output_size))
    return W1, b1, W2, b2, W3, b3

# Forward pass
def forward(X, W1, b1, W2, b2, W3, b3):
    Z1 = X @ W1 + b1
    A1 = ReLU(Z1)
    Z2 = A1 @ W2 + b2
    A2 = ReLU(Z2)
    Z3 = A2 @ W3 + b3
    A3 = softmax(Z3)
    return Z1, A1, Z2, A2, Z3, A3

# Backward pass
def backward(X, y, Z1, A1, Z2, A2, A3, W2, W3):
    m = X.shape[0]
    dZ3 = A3 - y
    dW3 = A2.T @ dZ3 / m
    db3 = np.sum(dZ3, axis=0, keepdims=True) / m

    dA2 = dZ3 @ W3.T
    dZ2 = dA2 * ReLU_derivative(Z2)
    dW2 = A1.T @ dZ2 / m
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    dA1 = dZ2 @ W2.T
    dZ1 = dA1 * ReLU_derivative(Z1)
    dW1 = X.T @ dZ1 / m
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    return dW1, db1, dW2, db2, dW3, db3

# Update weights
def update(W1, b1, dW1, db1, W2, b2, dW2, db2, W3, b3, dW3, db3, lr):
    W1 -= lr * dW1
    b1 -= lr * db1
    W2 -= lr * dW2
    b2 -= lr * db2
    W3 -= lr * dW3
    b3 -= lr * db3
    return W1, b1, W2, b2, W3, b3

# Training loop
def train(X, y, input_size, hidden1_size, hidden2_size, output_size, epochs, lr, lemda):
    W1, b1, W2, b2, W3, b3 = init_weights(input_size, hidden1_size, hidden2_size, output_size)
    loss_history = []
    m = X.shape[0]

    for i in range(epochs):
        Z1, A1, Z2, A2, Z3, A3 = forward(X, W1, b1, W2, b2, W3, b3)
        
        loss = -np.mean(np.sum(y * np.log(A3 + 1e-8), axis=1)) 
        reg = (lemda / (2 * m)) * (np.sum(W1**2) + np.sum(W2**2) + np.sum(W3**2))
        total_loss = loss + reg
        loss_history.append(total_loss)

        dW1, db1, dW2, db2, dW3, db3 = backward(X, y, Z1, A1, Z2, A2, A3, W2, W3)

        dW1 += (lemda / m) * W1
        dW2 += (lemda / m) * W2
        dW3 += (lemda / m) * W3

        W1, b1, W2, b2, W3, b3 = update(W1, b1, dW1, db1, W2, b2, dW2, db2, W3, b3, dW3, db3, lr)

    

    return W1, b1, W2, b2, W3, b3, loss_history

# Accuracy function
def accuracy(X, y, W1, b1, W2, b2, W3, b3):
    _, _, _, _, _, A3 = forward(X, W1, b1, W2, b2, W3, b3)
    preds = np.argmax(A3, axis=1)
    true = np.argmax(y, axis=1)
    return np.mean(preds == true) * 100

# Network configuration
input_size = X_train.shape[1]
hidden1_size = 100
hidden2_size = 50
output_size = y_train.shape[1]

# Train the model
W1, b1, W2, b2, W3, b3, loss_history = train(
    X_train, y_train,
    input_size=input_size,
    hidden1_size=hidden1_size,
    hidden2_size=hidden2_size,
    output_size=output_size,
    epochs=5000,
    lr=0.1,
    lemda=0.1
)

# Report
print("Final loss:", loss_history[-1])
print("Training accuracy: {:.2f}%".format(accuracy(X_train, y_train, W1, b1, W2, b2, W3, b3)))

# Plot the training loss
plt.figure(figsize=(8, 4))
plt.plot(loss_history, color='purple')
plt.title("Training Loss over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Cross-Entropy Loss")
plt.grid(True)
plt.show()
```
